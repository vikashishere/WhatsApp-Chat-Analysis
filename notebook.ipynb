```python
import numpy as np
import pandas as pd
import re
```

```python
f = open('WhatsApp Chat with Akash Jio.txt', 'r', encoding='utf-8')
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
# pattern = '\d{1,2}/\d{1,2}/\d{1,2},\s\d{1,2}:\d{2}\s-\s'
pattern = '\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{1,2}\s[am|pm]{2}\s-\s'
```

```python
messages = re.split(pattern, data)
```

```python
# First msg was coming as an empty string hence printing from 2nd onwards
messages[1:]
```

```python
messages = messages[1:]
```

```python
print(len(messages))
```

```python
dates = re.findall(pattern, data)
dates
```

```python
print(len(dates))
```

```python
df = pd.DataFrame({'user_message': messages, 'date': dates})

df.head()
```

```python
def remove_u202f(txt):
    txt = txt.replace("am", "")
    txt = txt.replace("pm", "")
    return txt
```

```python
df['date'] = df['date'].apply(remove_u202f)
df.head()
```

```python
# changing the type of msg_date column
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y, %H:%M - ')
df.head()
```

```python
df.shape
```

```python
#separating users and messages

users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]: #user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
        
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
# extracting year, month columns

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['min'] = df['date'].dt.minute
df.head()
```
